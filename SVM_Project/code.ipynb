```python
from sklearn import svm
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score 
from sklearn.preprocessing import StandardScaler , LabelEncoder 
from sklearn.model_selection import train_test_split
```

```python
data.head()
```

```python
# LabelEncoder = Convert to Categorical Data
gender = LabelEncoder()
data["Gender"] = gender.fit_transform(data["Gender"])
```

```python
data.head()
```

```python
# Cut = Ages to groups of ages ranges , supports pinning into an equal no of bins , prespecified array of bins 
bins = (-1,0,1,2,3,4,5)
health = ['malnourished', 'underweight', 'fit', 'slightly overweight', 'overweight', 'extremely overweight']
data["Index"] = pd.cut(data["Index"] , bins = bins , labels=health)
```

```python
# svc = best hyperplane

X = data.drop("Index", axis = 1)
y = data["Index"]
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
# StandardScaler = standardizes features by removing mean and scaling to unit variance
# if not used ACCURACY will be affected

s= StandardScaler()

X_train = s.fit_transform(X_train)
X_test = s.transform(X_test)
```

```python
# Fit = best match curvature of a given datapoint

clf = svm.SVC()
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, pred))
```

```python
confusion_matrix(y_test, pred)
```

```python
accuracy_score(y_test, pred)
```
